```python
print(__doc__)

# Author: Gael Varoquaux <gael dot varoquaux at normalesup dot org>
# License: BSD 3 clause

# Standard scientific Python imports
import matplotlib.pyplot as plt
%matplotlib inline

# Import datasets, classifiers and performance metrics
from sklearn import datasets, svm, metrics

# The digits dataset
digits = datasets.load_digits()

# The data that we are interested in is made of 8x8 images of digits, let's
# have a look at the first 3 images, stored in the `images` attribute of the
# dataset.  If we were working from image files, we could load them using
# pylab.imread.  Note that each image must have the same size. For these
# images, we know which digit they represent: it is given in the 'target' of
# the dataset.
images_and_labels = list(zip(digits.images, digits.target))
for index, (image, label) in enumerate(images_and_labels[:4]):
    plt.subplot(2, 4, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Training: %i' % label)

# To apply a classifier on this data, we need to flatten the image, to
# turn the data in a (samples, feature) matrix:
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))

# Create a classifier: a support vector classifier
classifier = svm.SVC(gamma=0.001)

# We learn the digits on the first half of the digits
classifier.fit(data[:n_samples / 2], digits.target[:n_samples / 2])

# Now predict the value of the digit on the second half:
expected = digits.target[n_samples / 2:]
predicted = classifier.predict(data[n_samples / 2:])

print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(expected, predicted)))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(expected, predicted))

images_and_predictions = list(zip(digits.images[n_samples / 2:], predicted))
for index, (image, prediction) in enumerate(images_and_predictions[:4]):
    plt.subplot(2, 4, index + 5)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Prediction: %i' % prediction)

plt.show()
```

#What we've learned so far.

##import
So we learned how to import a bunch of data stored in a single image sprite, using something like:
>  `digitsdata = ndimage.imread('data/digits.png', flatten=True, mode='L')`

Where each pixel is read in as a float value from 0-255 (greyscale). The shape of this numpy array is then 2000x1000 (shape of the original image in pixels). Each cell in this array is a digit from 0-255 representing intensity.
<BR>
##Splitting up the sprite image


Then we can break it up into individual images with a list comprehension:
>  `cells = [np.hsplit(row,100) for row in np.vsplit(digitsdata,50)]`

We know that each image is 20px square. The first thing we do is vsplit the 1000 vertical pixels into 50 chunks (ie. 1 chunk per 20 px). Each of these is called a "row" (list of sub-arrays). Then we split each row in 100 chunks (ie. 1 chunk per 20px). We now have a bunch of 20x20 cells

Each of these cells is then a 20x20 square. We can put these into a 4-dimensional numpy array *`X`*. Where x.shape = `(50, 100, 20, 20)`.

This means `x[0-50][0-100]` is now a 20x20 array of values.

```python

print np.shape(x)
print type(x)
X = x.reshape(5000,20,20)

print np.shape(X)
print np.shape(X[0])

#now, lets make sure we reshaped it all correctly!
for i in np.arange(1,5000,500):   #from 1 to 5000, in steps of 500
    plt.imshow(X[i])
    print "i is equal to: ", i
    plt.show()
```

```python
#borrowing from http://scikit-learn.org/stable/auto_examples/plot_classifier_comparison.html

from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.lda import LDA
from sklearn.qda import QDA

```

```python
X_train, X_test, y_train, y_test = train_test_split(X2,y, test_size=0.25, random_state=42)


names = ["Nearest Neighbors", "Linear SVM", "RBF SVM", "Decision Tree",
         "Random Forest", "AdaBoost", "LDA"]
classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    AdaBoostClassifier(),
    LDA()]
```

```python
print "Classifier\t\tScore"
for name, clf in zip(names, classifiers):
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)
        print "%s \t\t %.3f" % (name, score)
```

##Results:
Seems like KNN is the way to go. Easy, simple, and it's super easy to use. Lets see if we can improve our results at all via PCA or changing the number of neighbors

```python
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV
from sklearn.decomposition import PCA, RandomizedPCA
```

```python
pipeline = Pipeline([
#         ('pca', PCA()),
        ('knn', KNeighborsClassifier())
])

parameters = {'knn__n_neighbors': [2,3,4,5,6], 
#               'knn__weights': ['uniform', 'distance'],
              'knn__leaf_size': [20,30,40,50]}

gridsearch = GridSearchCV(pipeline,parameters, n_jobs=1, verbose=2, scoring='accuracy')
#CAUTION: for some reason this hangs when using anything other than n_jobs=1.
```

```python
gridsearch.fit(X_train, y_train)
```

```python
print gridsearch.score(X_test, y_test)
print gridsearch.best_estimator_
```

```python

def invertImageValues(original):
    # input is an array (1-D) of 0-255 values
    # returns the invert of that image (black -> white)
    
    k = np.array([255]* len(original))
    return abs(original - k)


def readPNG(filename):
    # input is a string for a filename
    # input MUST be a square image
    # return is a 1-D array
    
    rawdata = ndimage.imread(filename, flatten=True)
    #TODO: throw an exception if the image is not square
    
    len = np.shape(rawdata)[0]
    return np.reshape(rawdata, (len*len,))
    
    
testdata = ['data/test/3.png', 'data/test/6.png']
```
